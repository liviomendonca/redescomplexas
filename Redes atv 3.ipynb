# Bibliotecas

```python
import networkx as nx
import random
from random import *
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from scipy.stats import binom
from scipy.stats import poisson
import seaborn as sns
from math import factorial
```

# Funções

```python
def gerar_rede(N):
    G = nx.Graph() 
    G.add_nodes_from(range(1, N+1))

    fator = 50/N # pra garantir o grau médio
    print(fator)

    for i in range(1, N):
        for j in range(i+1, N+1):
            p = random()
            if p < fator:
                G.add_edge(i, j)

    nx.draw(G)
    return G
```

### Instruções

1. Start with N isolated nodes.
2. Select a node pair and generate a random number between 0 and 1.
3. If the number exceeds p, connect the selected node pair with a link, otherwise leave them disconnected.
4. Repeat step (2) for each of the N(N-1)/2 node pairs.

# Rede com 10^2 nós

```python
g1 = gerar_rede(100)
```

```python
grausg1 = [] # guardar só o grau de cada nó
N_kg1 = [] # qtd de nós com grau k
p_kg1 = [] # probabilidade

for i in g1.nodes:
    grausg1.append(g1.degree[i])

set_grausg1 = list(set(grausg1))

for i in range(len(set_grausg1)):
    N_kg1.append(grausg1.count(set_grausg1[i]))

for i in range(len(N_kg1)):
    p_kg1.append(N_kg1[i]/100)

k_mediog1 = sum(grausg1)/100

print(k_mediog1)
print(set_grausg1)
```

```python
X = binom.rvs(100, k_mediog1/100, size=100) # ajuste dos dados via binomial
Y = poisson.rvs(mu=k_mediog1, size=100) # ajuste dos dados via poisson

sns.displot((X, Y, grausg1), kind="kde")
```
