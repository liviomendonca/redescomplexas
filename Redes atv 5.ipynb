Assortativity and Pearson Correlation

```python
A = nx.read_edgelist('powergrid.edgelist.txt', create_using=nx.Graph(), nodetype = int)
```

```python
plt.figure(figsize=(7,7))
nx.draw_networkx(A, node_size=1, edge_color='grey', alpha=0.5, width=0.5, with_labels=False)
plt.show()
```

```python
print(nx.degree_assortativity_coefficient(A))
print(nx.degree_pearson_correlation_coefficient(A))
```
